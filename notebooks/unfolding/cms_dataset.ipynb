```python
import torch
import numpy as np
from experiments.unfolding.utils import get_pt
```

```python
gen_jets = np.load('data/cms/gen_1725_delphes.npy')
gen_jets = torch.from_numpy(gen_jets)
det_jets = np.load('data/cms/gen_1725_delphes.npy')
det_jets = torch.from_numpy(det_jets)
```

```python
gen_jet = gen_jets[..., :4]
gen_pt = get_pt(gen_jet)
det_jet = det_jets[..., :4]
det_pt = get_pt(det_jet)
```

```python
import matplotlib.pyplot as plt

# Assuming gen_pt and det_py are already defined
bins = np.linspace(0, 1000, 100)
plt.hist(gen_pt, bins=bins, alpha=0.5, label='gen_pt', histtype='step')
plt.hist(det_pt, bins=bins, alpha=0.5, label='det_pt', histtype='step')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.xlim(150, 600)
plt.title('Histogram of gen_pt and det_pt')
plt.legend(loc='upper right')
plt.show()
```
